```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
import seaborn as sb
import datetime as dt

df3 = pd.read_excel ('TSLA_2017_2020.xlsx')
df3 = df3.set_index(pd.DatetimeIndex(df3['Date'].values))
df3.head()
```

```python
short = 5
long = 10

df3['Close_short'] = df3['Close'].rolling(window=short).mean()
df3['Close_long'] = df3['Close'].rolling(window=long).mean()

df3
```

```python
close_position = df3.tail(17)
print(close_position)
```

```python
def strategy(close_position):
    status = 0
    PriceBuy = []
    PriceSell = []
    Revenue = []
    Temp_buyprice = -1
             
    for i in range (len(close_position)):
        if close_position['Close_short'][i] > close_position['Close_long'][i]:
            # Signal to buy
            if status != 1:
                status = 1
                PriceBuy.append(close_position['Open'][i+1])
                PriceSell.append(np.nan)
                Temp_buyprice = close_position['Open'][i+1]
                Revenue.append(np.nan)
            else:
                PriceBuy.append(np.nan)
                PriceSell.append(np.nan)
                Revenue.append(np.nan)
            if status == 1:
                for i in close_position['Open']:
                    Max_open= max(close_position['Open'])
                    break
                
                a = Max_open
                if a > 0:
                    print('sell today ' + str(a))
                    PriceSell.append(max(close_position['Open']))
                    Revenue.append(max(close_position['Open'])- Temp_buyprice)
                    break                    
        else:
            PriceBuy.append(np.nan)
            PriceSell.append(np.nan)
            Revenue.append(np.nan)
    return (PriceBuy, PriceSell, Revenue)            
```

```python
Revenuelastsell = pd.DataFrame((strategy(close_position)))
Revenuelastsell = Revenuelastsell.T
Revenuelastsell.columns = ['Price Buy', 'Price Sell','Revenue']
Revenuelastsell.to_excel('Revenuelastsell.xlsx')
```
